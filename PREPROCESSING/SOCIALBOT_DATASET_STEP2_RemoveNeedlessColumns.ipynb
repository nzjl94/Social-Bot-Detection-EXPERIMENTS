```python
import json
import csv
import numpy as np
import pandas as pd
from urllib.parse import quote  
```

```python
pd.set_option('display.max_columns', None)
pd.set_option("max_rows", None)
```

```python
data = pd.read_csv(f'PATH/t-step1.csv.gz',compression='gzip',lineterminator='\n');
```

```python
data.drop(columns=["lang","time_zone","follow_request_sent","utc_offset","notifications","following"], axis=1, inplace=True); # all values are NAN

data.drop(columns=["protected","contributors_enabled"], axis=1, inplace=True); # all values are False

# is_translation_enabled  ---> 1931 is true the remaining are false
# is_translator           ---> 16 is true the remaining are false
data.drop(columns=["is_translator","is_translation_enabled"], axis=1, inplace=True);
```

```python
# column_name,total_number, number_of_NAN
# profile_sidebar_fill_color 25876,0
# profile_link_color 42505,0
# profile_sidebar_border_color 14274,0
# profile_background_color 27784,0
# profile_text_color 22636,0


test= data.groupby('profile_sidebar_fill_color')["profile_sidebar_fill_color"].agg(["count"]).query("count > 1000").sort_values(by=['count'], ascending=False)
test.to_csv(f'PATH/t-profile_sidebar_fill_color.csv')
test.head(10)

test= data.groupby('profile_link_color')["profile_link_color"].agg(["count"]).query("count > 1000").sort_values(by=['count'], ascending=False)
test.to_csv(f'PATH/t-profile_link_color.csv')
test.head(10)

test= data.groupby('profile_sidebar_border_color')["profile_sidebar_border_color"].agg(["count"]).query("count > 1000").sort_values(by=['count'], ascending=False)
test.to_csv(f'PATH/t-profile_sidebar_border_color.csv')
test.head(10)

test= data.groupby('profile_background_color')["profile_background_color"].agg(["count"]).query("count > 1000").sort_values(by=['count'], ascending=False)
test.to_csv(f'PATH/t-profile_background_color.csv')
test.head(10)

test= data.groupby('profile_text_color')["profile_text_color"].agg(["count"]).query("count > 1000").sort_values(by=['count'], ascending=False)
test.to_csv(f'PATH/t-profile_text_color.csv')
test.head(10)

data.drop(columns=["profile_sidebar_fill_color","profile_link_color","profile_sidebar_border_color","profile_background_color","profile_text_color"], axis=1, inplace=True);
```

```python
# number of unique values
# id 1246936
# id_str 1246936
# user_created_at 1054849
# created_at 1240704

data.drop(columns=["id","id_str","user_created_at","created_at"], axis=1, inplace=True);
```

```python
# withheld_in_countries
# NaN       1246899
# ['TR']         26
# ['RU']          3
# ['IN']          3
# ['FR']          3
# ['DE']          2


# translator_type
# none         1240394
# regular         6526
# badged            14
# moderator          2

data.drop(columns=["withheld_in_countries","translator_type"], axis=1, inplace=True);
```

```python
#This line should have removed in the step 2
# we remove it becuase it is the same as [profile_background_image_url_https] column
data.drop(columns=["profile_background_image_url"], axis=1, inplace=True);
```

```python
# profile_image_url_https 1175034,20  --->>> it has the same value as [profile_image_url]
data.drop(columns=["profile_image_url_https"], axis=1, inplace=True);
```

```python
# [profile_image_url] it contains ether profile_image or default profile image
# [default_profile_image] this feature cover [profile_image_url] because of that we don't need this feature
data.drop(columns=["profile_image_url"], axis=1, inplace=True);
```
