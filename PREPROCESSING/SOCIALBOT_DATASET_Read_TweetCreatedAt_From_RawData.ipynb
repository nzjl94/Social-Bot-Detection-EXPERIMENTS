```python
import csv

archive = zipfile.ZipFile('PATH/CollectedData.zip', 'r')
file_object = open(f"PATH/createdAtTwitter.csv", 'a')
writer = csv.DictWriter(file_object, fieldnames=["id","created_at"])

for index,f in enumerate(reversed(archive.namelist())):
    if f.endswith("/"):
        pass    
    else:
        print(f)
        with archive.open(f,mode='r') as file:
            for line in file:
              data=json.loads(line.rstrip().decode('UTF-8'))
              writer.writerow({"id":data["user"]["id"],"created_at":data["created_at"]})

file_object.close()


with open(f"PATH/createdAtTwitter.csv", 'rb') as f_in:
  with gzip.open(f"PATH/createdAtTwitter.csv.gz", 'wb') as f_out:
    shutil.copyfileobj(f_in, f_out)


if os.path.exists(f"PATH/createdAtTwitter.csv"):
  os.remove(f"PATH/createdAtTwitter.csv")
```
