```python
tweets = pd.read_csv(f'PATH/CRESCI_2017/genuine_accounts/tweets.csv',encoding = "ISO-8859-1");
data =   pd.read_csv(f'PATH/CRESCI_2017/genuine_accounts/users.csv');
```

```python
# Below Rows needed for genuine_accounts
# data= data[(data["test_set_1"]==0) & (data["test_set_2"]==0)] --->>> last version ignore this condition to increase sample number
data.drop(columns=['timestamp', 'crawled_at','test_set_1', 'test_set_2'], axis=1, inplace=True);

# Below Rows needed for fake_followers
data=data[A]

# Below Rows needed for social_spambots_1
# data["test_set_1"].value_counts()
data.drop(columns=['timestamp', 'crawled_at','test_set_1'], axis=1, inplace=True);

# Below Rows needed for social_spambots_2
data.drop(columns=['timestamp', 'crawled_at'], axis=1, inplace=True);

# Below Rows needed for social_spambots_3
data.drop(columns=['timestamp', 'crawled_at','test_set_2'], axis=1, inplace=True);

# Below Rows needed for traditional_spambots_1
data.drop(columns=['timestamp', 'crawled_at'], axis=1, inplace=True);
```

```python
tweets=tweets.sort_values('created_at', ascending=False).drop_duplicates(['user_id'])
tweets=tweets[['user_id','created_at']]
tweets.rename({'created_at': 'tweets_created_at','user_id': 'id'}, axis=1,inplace=True)
final_result=pd.merge(data,tweets,on='id',how='left')


Some file need those to convert dae
final_result["created_at"] = pd.to_numeric(final_result["created_at"].str.replace('L', ''))
final_result["tweets_created_at"] = pd.to_numeric(final_result["tweets_created_at"].str.replace('L', ''))
final_result['diff_days'] = (pd.to_datetime(final_result['tweets_created_at'],utc=True, unit='ms')-pd.to_datetime(final_result['created_at'],utc=True, unit='ms')).dt.days

# Some other file need those to convert dae
final_result['diff_days'] = (pd.to_datetime(final_result['tweets_created_at'])-pd.to_datetime(final_result['created_at'])).dt.days


final_result['diff_days'].replace(0, 1,inplace=True)
final_result["diff_days"]=np.where(final_result["diff_days"].isnull(),1,final_result["diff_days"])
```
