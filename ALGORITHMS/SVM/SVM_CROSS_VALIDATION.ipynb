```python
import json
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import KFold
from sklearn.svm import SVC 
from sklearn.model_selection import cross_val_score
```

```python
pd.set_option('display.max_columns', None)
pd.set_option("max_rows", None)
```

```python
data = pd.read_csv(f'PATH/implementation_sample_1.csv.gz',compression='gzip',lineterminator='\n');
```

```python
y = data['class'].to_numpy()
X = data.drop(['class'],axis=1).to_numpy()
svc_model = SVC()


kf = KFold(n_splits=5, shuffle=True, random_state=1111)
splits = kf.split(X)

for train_index, val_index in splits:
    X_train, y_train = X[train_index], y[train_index]
    X_val, y_val = X[val_index], y[val_index]
    
    svc_model.fit(X_train, y_train)
    
    predictions = svc_model.predict(X_val)
    print(classification_report(y_val, predictions))
```

```python
cv10 = cross_val_score(svc_model, X, y, cv=3, scoring='accuracy')
print("Ten-fold cross validation:", cv10)
```
