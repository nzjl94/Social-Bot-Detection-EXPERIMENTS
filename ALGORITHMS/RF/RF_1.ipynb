```python
import json
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix,ConfusionMatrixDisplay
from sklearn.ensemble import RandomForestClassifier
```

```python
pd.set_option('display.max_columns', None)
pd.set_option("max_rows", None)
```

```python
data = pd.read_csv(f'PATH/DS1.csv.gz',compression='gzip',lineterminator='\n');
data=data.sample(frac=1).reset_index(drop=True)
```

```python
y = data['class']
X = data.drop(['class'],axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

```python
randomforest_classifier = RandomForestClassifier(n_estimators = 100, criterion = 'entropy')
randomforest_classifier.fit(X, y)
```

```python
type(classification_report(y, randomforest_classifier.predict(X),output_dict=True))
```

```python
from sklearn.ensemble import RandomForestClassifier

randomforest_classifier = RandomForestClassifier(n_estimators = 100, criterion = 'entropy')
randomforest_classifier.fit(X, y)

y_predict_train = randomforest_classifier.predict(X_train)
cm = confusion_matrix(y_train, y_predict_train)
sns.heatmap(cm, annot=True,fmt="d")
```

```python
print(classification_report(y_train, y_predict_train))
```

```python
y_predict = randomforest_classifier.predict(X_test)
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot=True,fmt="d")
```

```python
print(classification_report(y_test, y_predict))
```
