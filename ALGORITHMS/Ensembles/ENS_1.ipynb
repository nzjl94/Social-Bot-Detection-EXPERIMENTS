```python
def get_voting():
	models = list()
	models.append(('adaboost',AdaBoostClassifier()))
	models.append(('bagging',BaggingClassifier(n_estimators=100)))
	models.append(('randomForest',RandomForestClassifier(n_estimators=100)))
	models.append(('extraTrees',ExtraTreesClassifier(n_estimators=100)))


	ensemble = VotingClassifier(estimators=models, voting='soft')
	return ensemble
```
