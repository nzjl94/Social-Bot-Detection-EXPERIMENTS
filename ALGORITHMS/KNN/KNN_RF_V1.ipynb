```python
import json
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
%matplotlib inline
```

```python
pd.set_option('display.max_columns', None)
pd.set_option("max_rows", None)
```

```python
data = pd.read_csv(f'PATH/implementation_sample_2.csv.gz',compression='gzip',lineterminator='\n');
```

```python
DS1=pd.concat([data[data["class"]=="bot"],data[data["class"]=="human"].sample(60364)]);
DS1 = DS1.sample(frac=1).reset_index(drop=True)
DS2=pd.concat([data[data["class"]=="bot"],data[data["class"]=="human"].sample(60364)]);
DS2 = DS2.sample(frac=1).reset_index(drop=True)
DS3=pd.concat([data[data["class"]=="bot"],data[data["class"]=="human"].sample(60364)]);
DS3 = DS3.sample(frac=1).reset_index(drop=True)
DS4=pd.concat([data[data["class"]=="bot"],data[data["class"]=="human"].sample(60364)]);
DS4 = DS4.sample(frac=1).reset_index(drop=True)
DS5=pd.concat([data[data["class"]=="bot"],data[data["class"]=="human"].sample(60364)]);
DS5 = DS5.sample(frac=1).reset_index(drop=True)
DS6=pd.concat([data[data["class"]=="bot"],data[data["class"]=="human"].sample(60364)]);
DS6 = DS6.sample(frac=1).reset_index(drop=True)
DS7=pd.concat([data[data["class"]=="bot"],data[data["class"]=="human"].sample(60364)]);
DS7 = DS7.sample(frac=1).reset_index(drop=True)
```

```python
y_1 = DS1['class']
X_1 = DS1.drop(['class'],axis=1)
X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(X_1, y_1, test_size=0.1)

y_2 = DS2['class']
X_2 = DS2.drop(['class'],axis=1)
X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(X_2, y_2, test_size=0.1)

y_3 = DS3['class']
X_3 = DS3.drop(['class'],axis=1)
X_train_3, X_test_3, y_train_3, y_test_3= train_test_split(X_3, y_3, test_size=0.1)

y_4 = DS4['class']
X_4 = DS4.drop(['class'],axis=1)
X_train_4, X_test_4, y_train_4, y_test_4 = train_test_split(X_4, y_4, test_size=0.1)

y_5 = DS5['class']
X_5 = DS5.drop(['class'],axis=1)
X_train_5, X_test_5, y_train_5, y_test_5 = train_test_split(X_5, y_5, test_size=0.1)

y_6 = DS6['class']
X_6 = DS6.drop(['class'],axis=1)
X_train_6, X_test_6, y_train_6, y_test_6 = train_test_split(X_6, y_6, test_size=0.1)

y_7 = DS7['class']
X_7 = DS7.drop(['class'],axis=1)
X_train_7, X_test_7, y_train_7, y_test_7 = train_test_split(X_7, y_7, test_size=0.1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier


RF_1 = RandomForestClassifier(n_estimators = 100, criterion = 'entropy')
RF_2 = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
RF_3 = RandomForestClassifier(n_estimators = 100, criterion = 'entropy')
RF_4 = KNeighborsClassifier(n_neighbors = 7, metric = 'minkowski', p = 2)
RF_5 = RandomForestClassifier(n_estimators = 100, criterion = 'entropy')
RF_6 = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
RF_7 = RandomForestClassifier(n_estimators = 100, criterion = 'entropy')

RF_1.fit(X_train_1, y_train_1)
RF_2.fit(X_train_2, y_train_2)
RF_3.fit(X_train_3, y_train_3)
RF_4.fit(X_train_4, y_train_4)
RF_5.fit(X_train_5, y_train_5)
RF_6.fit(X_train_6, y_train_6)
RF_7.fit(X_train_7, y_train_7)
```

```python
y_predict_train_1 = RF_1.predict(X_train_1)
print(classification_report(y_train_1, y_predict_train_1))
cm = confusion_matrix(y_train_1, y_predict_train_1)
sns.heatmap(cm, annot=True,fmt="d")
```
