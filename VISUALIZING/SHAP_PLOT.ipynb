```python
import json
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix,ConfusionMatrixDisplay
from sklearn.ensemble import RandomForestClassifier
from xgboost                    import XGBClassifier
import shap 
from sklearn.preprocessing      import QuantileTransformer
```

```python
pd.set_option('display.max_columns', None)
pd.set_option("max_rows", None)
```

```python
data = pd.read_csv(f'PATH/DS1.csv.gz',compression='gzip',lineterminator='\n');
data=data.sample(frac=1).reset_index(drop=True)

data.rename(columns={
    'diff_days':'account_age',
    'entities':'has_entities',
    '_location':'has_location',
    '_url':'has_url'
}, inplace=True)
```

```python
normalizingModels=QuantileTransformer(output_distribution="normal",n_quantiles=len(data))

columns=data.drop(['class'],axis=1).columns

X = normalizingModels.fit_transform(data.drop(['class'],axis=1))
y = data['class'];
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.1)
```

```python
class_1 = XGBClassifier()
class_1.fit(X_train, y_train)
explainer = shap.TreeExplainer(class_1)
shap_values = explainer.shap_values(X_test)
```

```python
shap.summary_plot(shap_values, X_test,show=True, max_display=25,feature_names=columns)
```

```python
class_2 = RandomForestClassifier(n_estimators = 100)
class_2.fit(X_train, y_train)
explainer = shap.TreeExplainer(class_2)
shap_values = explainer.shap_values(X_test)
```

```python
shap.summary_plot(shap_values[1], X_test,show=True, max_display=25,feature_names=columns)
```
