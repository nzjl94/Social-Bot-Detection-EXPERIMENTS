```python
fig, axs = plt.subplots(2,6, figsize=(30, 10))


datasets=[
    ["DS4","DS5","DS6","DS7","DS12","DS13"],
    ["DS14","DS15","DS16","DS20","DS21"]
]
datasetsLabel=[
    ["Cresci-rtbust","Cresci-stock","Gilani-17","Midterm-18","Pron-celebrity","Political-feedback"],
    ["Botwiki-verified","Vendor-verified","Cresci-17","Pro1","Pro2"]
]
SelectedFeatures=[ "favourites_count","friends_growth_rate","friends_count","followers_count","favourites_growth_rate","tweet_freq","num_digits_in_screen_name","followers_growth_rate","followers_friends_ratio","description_length","class"]



def calculaing_data(dataset,dataset_name):

    DS = pd.read_csv(f'/PATH/{dataset}.csv.gz',compression='gzip',lineterminator='\n');

    print(DS.columns)

    DS=DS[SelectedFeatures]

    IS_HUMAN=DS["class"]=="human"
    IS_BOT=DS["class"]=="bot"
    total_number=500

    if len(DS[IS_HUMAN])>total_number and len(DS[IS_BOT])>total_number:
        DS=pd.concat([DS[IS_BOT].sample(total_number),DS[IS_HUMAN].sample(total_number)])
    elif len(DS[IS_HUMAN])>total_number:
        DS1=pd.concat([DS[IS_BOT],DS[IS_HUMAN].sample(len(DS[IS_BOT]))])
    elif len(DS[IS_BOT])>total_number:
        DS=pd.concat([DS[IS_BOT].sample(len(DS[IS_HUMAN])),DS[IS_HUMAN]])

    
    DS=DS.sample(frac=1).reset_index(drop=True)

    Y=DS["class"]    
    X = Normalizer().fit_transform(DS.drop(columns=["class"],axis=1))


    TSNE_RESULT = TSNE(n_components=2,init="pca").fit_transform(X)

    Final_RESULT = pd.DataFrame({
        'tsne_1': TSNE_RESULT[:,0], 
        dataset_name: TSNE_RESULT[:,1], 
        'label': Y
    })

    return Final_RESULT




for index_row,datasets_list in enumerate(datasets):
  for index_col,dataset in enumerate(datasets_list):

        Final_RESULT=calculaing_data(dataset,datasetsLabel[index_row][index_col])
        g=sns.scatterplot(y='tsne_1', x=datasetsLabel[index_row][index_col], data=Final_RESULT, ax=axs[index_row][index_col],hue='label',palette={"bot":"red","human":"blue"},s=100)

        axs[index_row][index_col].set_ylabel("")

        if index_row==0:
            axs[index_row][index_col].xaxis.set_label_position('top') 
        
        if index_row==1 and index_col==4:
            axs[index_row][index_col].legend(bbox_to_anchor=(1.58, 1.04), loc="upper right",fontsize='xx-large')
        else:
            axs[index_row][index_col].get_legend().remove()

axs[1][5].axis('off')

plt.show()
```
